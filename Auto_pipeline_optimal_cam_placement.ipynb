```python
def triangulate_fireNum(fireNum):
    overlap_fire_footage = fire_loc[fire_loc['Fire#']==('Fire'+str(fireNum))]
    overlap_fire_footage
    angle_info = []
    for index, row  in overlap_fire_footage.iterrows():
        #print(int(row['camera#']))
        #print(int(row['xmin']))
        #print(int(row['xmax']))
        X_orientation = (row['xmin'] + row['xmax'])/2
        camera_info = camera_video1_loc[camera_video1_loc['camera#']==int(row['camera#'])]
        X_resolution = 1920
        #print(camera_info)
        hFOV = camera_info['hFOV']
        
        angle = (X_orientation - X_resolution/2)/X_resolution * hFOV
        #print(int(camera_info['hFOV']))
        #print("angle", angle, int(camera_info['X_raster_loc']), camera_info['Y_raster_loc'])
        angle_info.append([angle, int(camera_info['X_raster_loc']), int(camera_info['Y_raster_loc'])])
        #print(cam_num)
        
    numCams = len(angle_info)
    X_list = []
    Y_list = []
    for i in range(numCams):
        for j in range(i + 1, numCams):
            alpha_1, X_1, Y_1 = angle_info[i]
            alpha_2, X_2, Y_2 = angle_info[j]
            print(alpha_1, alpha_2)
            #alpha  = 270 + 15.136902760416667
            #beta = 1
            alpha_1 = math.radians(alpha_1)#
            alpha_2 = math.radians(alpha_2) #
            X_3 = ((Y_1 - Y_2) + X_2 * math.tan(alpha_2) - X_1 *math.tan(alpha_1))/ (math.tan(alpha_2) - math.tan(alpha_1))
            Y_3 = ((Y_1 * math.tan(alpha_2) - Y_2 * math.tan(alpha_1)) - ((X_1 - X_2) * math.tan(alpha_2) * math.tan(alpha_1)))/ (math.tan(alpha_2) - math.tan(alpha_1))
            X_list.append(X_3)
            Y_list.append(Y_3)
    X_average = sum(X_list) / len(X_list)
    Y_average = sum(Y_list) / len(Y_list)
    return X_average, Y_average
```

```python
def DeepTopoMono(X_pixel, Y_pixel, X_raster, Y_raster, vFOV, hFOV, pitch, elevation, distance):
    #calculate the distance that is parallel to the 0 plane surface
    horizontal_dist = distance * math.cos(math.radians(vFOV * (Y_pixel - 1080/2)/1080))
    #calculate the distance of the pixel
    u_dist = horizontal_dist * math.cos(math.radians(hFOV * (X_pixel - 1920/2)/1920))
    v_dist = horizontal_dist * math.sin(math.radians(hFOV * (X_pixel - 1920/2)/1920))
    X_pred = X_raster - u_dist
    Y_pred = Y_raster - v_dist
    return X_pred, Y_pred
```

```python
def topo_Depth_C_terrain(X_pixel, Y_pixel, X_raster, Y_raster, vFOV, hFOV, pitch, elevation, terrain_model):

    elev_cam = elevation_calc(elevation)
    k_r = raster_k(X_pixel,vFOV)
    k_b = beam_k_p(Y_pixel, hFOV, pitch)
    d_b = elev_cam - k_b*X_raster
    d_r = Y_raster - k_r*X_raster
    for i in range(0, 1200):
        x = math.floor(X_raster - i)
        y = math.floor(k_r*x+d_r)
        view_elevation = k_b * x + d_b
        #print(x, y, img[y][x])
        #print(x, y, UE_selected[y][x],view_elevation)
        #elevation_array.append(img[y][x])
        #number_bin.append(x)
        #elevation_array.append(UE_selected[y][x])
        #number_bin.append(x)
        if (view_elevation < terrain_model[y][x]):      
            print(x, y, terrain_model[y][x],view_elevation)
            return x,y
    return -1, -1
```
